```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
sns.set_style('darkgrid')
```

```python
df = pd.read_csv(r"C:\Users\HP\Data Analysis Projects\Finance Analysis\Finance_data.csv")
```

```python
df.head()
```

```python
df.drop(['Mutual_Funds','Equity_Market','Debentures','Government_Bonds','Fixed_Deposits','PPF','Gold'], axis = 1, inplace = True)
```

```python
df = df.replace(r'^\s*$', np.nan, regex = True)
```

```python
sns.countplot(x='gender',hue='gender',data=df,linewidth=3,palette='Set2',edgecolor='black',legend=False)
plt.show()
```

```python
sns.countplot(x = 'age', hue='age',data = df, palette = "Set3", linewidth = 2, edgecolor = "black",legend=False)
plt.show()
```

```python
plt.figure(figsize =(18,6))
plt.subplot(1,2,1)
sns.countplot(x = df['gender'], hue = df['Investment_Avenues'], palette = 'summer', linewidth = 3, edgecolor = "white")
plt.title("INVESTMENT AVENUE")
plt.subplot(1,2,2)
sns.countplot(x = df['gender'], hue = df['Stock_Marktet'], palette = 'hot', linewidth = 3, edgecolor = 'white')
plt.title("STOCK MARKET")
plt.show()
```

```python
sns.countplot(x='Factor',hue='Factor',data=df,palette='coolwarm',linewidth=2,edgecolor="black",legend=False)
plt.show()
```

```python
plt.figure(figsize = (16,8))
sns.pointplot(x = "Purpose", y = "age", data = df, linestyles = "--", capsize = .3, color = "Blue")
plt.show()
```

```python
plt.figure(figsize = (18,6))
plt.subplot(1,2,1)
sns.countplot(hue = df['gender'], x = df['Duration'], palette = 'viridis', linewidth = 2, edgecolor = "black")
plt.title("DURATION")
plt.subplot(1,2,2)
sns.countplot(hue = df['gender'], x = df['Invest_Monitor'], palette = "seismic", linewidth = 2, edgecolor = "black")
plt.title("INVESTMENT MONITORING")
plt.show()
```
